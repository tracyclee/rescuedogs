```python
small_sub = small_df[small_df['animalOrgID']==1143]
dates = small_sub[small_sub['animalFoundDate']!='nothing']['animalFoundDate']
print dates
indices = dates.index
print len(dates)
index_pairs = [(i,j) for i, j in zip(indices[:-1], indices[1:])]
print index_pairs
date_pairs = np.array([(i,j) for i, j in zip(dates[:-1], dates[1:])])
print date_pairs
diffs = np.array([(j-i).days for i, j in zip(dates[:-1], dates[1:])])
print diffs
```

```python
test = 15350
which_pair = np.array([((test < pair[1]) and (test > pair[0])) for pair in date_pairs])
diffs[which_pair][0]
```

```python
small_sub['date_diff'] = small_sub['level_0'].apply(lambda x: 9999 if (x<indices[0] or x>indices[-1]) else diffs[np.array([((x <= pair[1]) and (x >= pair[0])) for pair in index_pairs])][0])
```

```python
keep_df = small_sub[abs(small_sub['date_diff'])<=30]
keep_df
```

```python
def label(row):
    if row['animalFoundDate']!='nothing':
        return row['animalFoundDate']
    else:
        bounds = date_pairs[np.array([((row['level_0']<pair[1]) and (row['level_0']>=pair[0])) for pair in index_pairs])][0]
        return bounds[0] + (bounds[1]-bounds[0])/2
    
```

```python
keep_df['imputed'] = keep_df.apply(lambda x: label(x), axis=1)
```
